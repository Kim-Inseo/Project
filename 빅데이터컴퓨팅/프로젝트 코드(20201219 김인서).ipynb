```python
data = [dust_2014, dust_2015, dust_2016, dust_2017, dust_2018, dust_2019, dust_2020, dust_2021, dust_2022, dust_2023]

dust = pd.concat(data, ignore_index=True)
dust
```

```python
dust.info()
```

```python
dust = dust.astype({'측정일자': 'str'})
```

```python
dust.info()
```

```python
year = []
for idx in range(len(dust)):
    year.append(dust.iloc[idx, 0][:4])
```

```python
month = []
for idx in range(len(dust)):
    month.append(dust.iloc[idx, 0][4:6])
```

```python
yyyymm = []
for idx in range(len(dust)):
    yyyymm.append(dust.iloc[idx, 0][:6])
```

```python
dust['연도'] = year
dust
```

```python
dust['월'] = month
dust
```

```python
dust['연월'] = yyyymm
dust
```
